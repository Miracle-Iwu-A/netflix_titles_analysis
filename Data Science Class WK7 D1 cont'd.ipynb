# Data Visualisation

- Visual representation of data to discover trends and pattern
- Helps in storytelling and decicion-making

### Types
- Categorical: Bar chart, pie chart, etc
- Numerical: Box plot, Histogram, etc
- Relationship: Satter plot, line plot, Heatmap, Pair plot, etc

### Libraries
- Matplotlib
- Seaborn

```python
import pandas as pd

import matplotlib.pyplot as plt
```

- Figure: The entire canvas that holds all subplots
- Axes: A single plot

```python
# Create a Figure with a single axes

x = [1, 2, 3]

y = [4, 5, 6]

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Basic Plot") # Use set_title to set the title of your plot
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

plt.show() # To disply the plot
```

```python
# create a figure with multiple subplots

fig, axes = plt.subplots(1, 2) #two plots side by side
axes[0].plot(x, y)
axes[0].set_title ('Left PLot')

axes[1].plot(x, y)
axes[1].set_title ('Right Plot')

plt.show()
```

```python
fie, axes = plt.subplots(2, 2)

axes[0, 0].plot(x, y)
axes[0, 1].plot(x, y)
axes[1, 0].plot(x, y)
axes[1, 1].plot(x, y)

plt.show()
```
